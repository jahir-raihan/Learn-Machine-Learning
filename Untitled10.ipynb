```python
# Dot product
# For 2-D vectors, it is the equivalent to matrix multplication
u = np.array([10, 20])
v = np.array([25, 6])

print(10 * 25 + 20 * 6)

np.dot(u, v)
```

```python
# Matrix and vector multiplication
a = [[5, 10], [15, 20]]
b = [10, 12]

np.dot(a, b)
```
