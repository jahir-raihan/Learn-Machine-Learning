```python
# Import the dependency for credit card fraud detection

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loading the data set

credit_card_data = pd.read_csv('creditcard.csv')
```

```python
# First five rows of the dataset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# dataset information

credit_card_data.info()
```

```python
# Checking the number of missing values in each columns

credit_card_data.isnull().sum()
```

```python
# Distribution of legit transaction and fraudlant transaction

credit_card_data['Class'].value_counts()
```

```python
# This dataset is hgihly unbalanced

# -- 0 Normal transaction
# -- 1 Fraud

# Separating the data for analysis

legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
legit.shape, fraud.shape
```

```python
# Statistical measures of the data

legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values for both transactions

credit_card_data.groupby('Class').mean()
```

```python
# Under-Sampling

# Build a sample dataset containing similar distributaion of normal transactions and
# Fraudulent transactions

legit_sample = legit.sample(n=492)
```

```python
# Let's concatenate two dataframes

new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# Splitting the data into features and targets

X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']

# train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
X_train.shape, X_test.shape, X.shape
```

```python
# Let's init the model

model = LogisticRegression()
```

```python
# Fit the data

model.fit(X_train, Y_train)
```

```python
# LEt's see the accuray

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
training_data_accuracy
```

```python
# Test data accuracy

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```
